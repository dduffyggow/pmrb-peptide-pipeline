```python
##Contacts ExPASy, requests a random protein sequence, and returns the Swiss-Prot text entry:

import requests
from bs4 import BeautifulSoup

def generate_randseq(length=50):                            # ðŸ”¶ CHANGE THIS (sequence length)
    url = "https://web.expasy.org/cgi-bin/randseq/randseq.pl"
    data = {
        "size": str(length),                                # ðŸ”¶ Set by the length argument above
        "composition": "average"                            # ðŸ”¶ CHANGE THIS if you want equal/user-defined/specific sequence
    }

    r = requests.post(url, data=data)

    # TEMPORARY: print response
    print(r.text)

    soup = BeautifulSoup(r.text, "html.parser")
    seq_block = soup.find("pre")

    if not seq_block:
        print("Could not find sequence in response.")
        return None

    return seq_block.text.strip()

test_seq = generate_randseq(50)
print(test_seq)
```

```python
##Takes the Swiss-Prot text, extracts the amino-acid sequence + accession, and saves a clean FASTA file.

def swissprot_to_fasta(text, outname="test_randseq_1.fasta"):            # ðŸ”¶ CHANGE THIS (output FASTA filename)
    # Split into lines
    lines = text.splitlines()

    # Extract accession number (AC   RNDxxxxx;)
    ac_line = [l for l in lines if l.startswith("AC")][0]
    accession = ac_line.split()[1].replace(";", "")

    # Extract sequence lines between SQ and //
    seq_lines = []
    capture = False
    for line in lines:
        if line.startswith("SQ"):
            capture = True
            continue
        if capture:
            if line.startswith("//"):
                break
            seq_lines.append(line.strip())

    # Join and remove spaces
    sequence = "".join(seq_lines).replace(" ", "")

    # Build FASTA string
    fasta = f">{accession}\n{sequence}"

    # Save it
    with open(outname, "w") as f:
        f.write(fasta + "\n")

    print(f"Saved FASTA as {outname}")
    return fasta
```

```python
##Generates one sequence, prints the Swiss-Prot output, converts it to FASTA, and saves the output file.

text_output = generate_randseq(50)           # ðŸ”¶ CHANGE THIS (sequence length)
print(text_output)

fasta_seq = swissprot_to_fasta(text_output, outname="test_randseq_1.fasta")             # ðŸ”¶ CHANGE THIS (output file name)
print(fasta_seq)
```

```python
import os
import requests
from bs4 import BeautifulSoup

def generate_randseq(length=20):                            # ðŸ”¶ 20 AA sequences
    url = "https://web.expasy.org/cgi-bin/randseq/randseq.pl"
    data = {
        "size": str(length),                                    # ðŸ”¶ This uses the length above
        "composition": "average"   # default mode             # ðŸ”¶ CHANGE IF YOU WANT: "average", "equal", "specific", "user"
    }

    r = requests.post(url, data=data)
    soup = BeautifulSoup(r.text, "html.parser")
    seq_block = soup.find("pre")

    if not seq_block:
        print("Could not find sequence in response.")
        return None

    return seq_block.text.strip()   # Return Swiss-Prot text
```

```python
def swissprot_to_fasta(text, outname):
    # Split into lines
    lines = text.splitlines()

    # Extract accession number
    ac_line = [l for l in lines if l.startswith("AC")][0]
    accession = ac_line.split()[1].replace(";", "")

    # Extract sequence lines between SQ and //
    seq_lines = []
    capture = False
    for line in lines:
        if line.startswith("SQ"):
            capture = True
            continue
        if capture:
            if line.startswith("//"):
                break
            seq_lines.append(line.strip())

    # Remove spaces and join
    sequence = "".join(seq_lines).replace(" ", "")

    # Build FASTA text
    fasta = f">{accession}\n{sequence}"

    # Save
    with open(outname, "w") as f:
        f.write(fasta + "\n")

    print(f"Saved FASTA: {outname}")
    return fasta
```

```python
# Create output folder
output_folder = "randseq_expasy_output"                    # ðŸ”¶ CHANGE THIS: name of output folder
os.makedirs(output_folder, exist_ok=True)

# Generate 10 sequences, each 20 AA
for i in range(1, 11):                                   # ðŸ”¶ CHANGE THIS: how many sequences to generate
    print(f"\n--- Generating sequence {i} ---")

    # Generate Swiss-Prot entry
    text_output = generate_randseq(20)                   # ðŸ”¶ CHANGE THIS: AA length for each sequence

    # Build filename inside output folder
    outname = os.path.join(output_folder, f"randseq_{i}.fasta")    # ðŸ”¶ CHANGE THIS IF YOU WANT CUSTOM FILENAMES

    # Convert and save
    fasta = swissprot_to_fasta(text_output, outname)
```

```python
## this produuced 10 separete fasta files... now let's merge this into one!...
```

```python
import os

output_folder = "randseq_expasy_output"                    # ðŸ”¶ CHANGE THIS if your folder name changes
merged_filename = os.path.join(output_folder, "randseq_merged.fasta")

with open(merged_filename, "w") as outfile:
    for fname in sorted(os.listdir(output_folder)):
        if fname.endswith(".fasta"):
            file_path = os.path.join(output_folder, fname)
            with open(file_path) as infile:
                # Strip trailing newlines but DO NOT add extra blank lines
                outfile.write(infile.read().strip() + "\n")
                
print(f"Merged FASTA saved as: {merged_filename}")
```
