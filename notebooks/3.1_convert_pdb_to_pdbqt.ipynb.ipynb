```python
# Convert PmrB receptor .pdb to .pdbqt using OpenBabel
import os
import subprocess

# Define paths
input_pdb = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/output_test/best_pmrb/UDY21468.1_PmrB__Acinetobacter_baumannii__unrelaxed_alphafold2_ptm_model_5_seed_000.pdb"
output_folder = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/converted_pdbqt/receptor_test"

# Make sure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Define output filename (simplified)
output_pdbqt = os.path.join(output_folder, "pmrb.pdbqt")

# Run OpenBabel conversion
try:
    subprocess.run(["obabel", input_pdb, "-O", output_pdbqt], check=True)
    print(f"\n✅ Successfully converted:\n{input_pdb}\n→ {output_pdbqt}")
except subprocess.CalledProcessError as e:
    print(f"\n❌ Error converting file: {e}")


# this was just PmrB 
```

```python
# Convert all Dcr ligand .pdb files to .pdbqt using OpenBabel
import os
import subprocess

# Define paths
ligands_folder = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/output_ligands/best_ligands"
output_folder = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/converted_pdbqt/ligands_test"

# Create the output directory
os.makedirs(output_folder, exist_ok=True)

# Loop through each ligand .pdb and convert
for file in os.listdir(ligands_folder):
    if file.endswith(".pdb"):
        ligand_name = file.split("_")[0]  # Extracts "Dcr1", "Dcr2", etc.
        input_path = os.path.join(ligands_folder, file)
        output_path = os.path.join(output_folder, f"{ligand_name}.pdbqt")

        try:
            subprocess.run(["obabel", input_path, "-O", output_path], check=True)
            print(f"✅ Successfully converted: {file} → {ligand_name}.pdbqt")
        except subprocess.CalledProcessError as e:
            print(f"❌ Error converting {file}: {e}")

print("\nAll ligand conversions complete!")
```
