```python
# Rank and copy best PmrB model (ColabFold output)
import os, json, numpy as np, shutil

# Path to your ColabFold output for PmrB
folder = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/output_test"
best_folder = os.path.join(folder, "best_pmrb")

# Create destination folder if it doesn't exist
os.makedirs(best_folder, exist_ok=True)

results = []  # store all (json_file, mean_plddt)

# Loop through JSON files and extract mean pLDDT
for file in os.listdir(folder):
    if file.endswith(".json") and "scores" in file:
        file_path = os.path.join(folder, file)
        with open(file_path) as f:
            data = json.load(f)
        if "plddt" in data:
            mean_plddt = np.mean(data["plddt"])
            results.append((file, mean_plddt))

# Sort by confidence (highest first)
results.sort(key=lambda x: x[1], reverse=True)

# Print all ranked models
for file, score in results:
    print(f"{file}: mean pLDDT = {score:.2f}")

# Copy best-scoring .pdb file to best_pmrb/
if results:
    best_json, best_score = results[0]
    best_pdb = best_json.replace(".json", ".pdb").replace("_scores_", "_unrelaxed_")
    pdb_path = os.path.join(folder, best_pdb)
    if os.path.exists(pdb_path):
        shutil.copy(pdb_path, best_folder)
        print(f"\n✅ Copied best PmrB model ({best_score:.2f}) → {best_folder}")
    else:
        print(f"⚠️ No matching PDB found for {best_json}")
else:
    print("❌ No valid JSON files found.")

print("\nAll top model processing complete.")
```

```python
# Extended version: Rank and copy best ligand models (ColabFold outputs)
import os, json, numpy as np, shutil

# Path to your ColabFold output folder
folder = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/output_ligands"
best_folder = os.path.join(folder, "best_ligands")

# Create the destination folder if it doesn’t exist
os.makedirs(best_folder, exist_ok=True)

results = {}       # Store all ranked models per ligand
best_models = {}   # Store only top models per ligand

# Loop through all files in the output folder
for file in os.listdir(folder):
    if file.endswith(".json") and "rank" in file and file.startswith("Dcr"):
        ligand = file.split("_")[0]  # Extract e.g. "Dcr1"
        file_path = os.path.join(folder, file)
        with open(file_path) as f:
            data = json.load(f)
            if "plddt" in data:
                mean_plddt = np.mean(data["plddt"])
                results.setdefault(ligand, []).append((file, mean_plddt))

# Sort and pick best model per ligand
for ligand, models in results.items():
    models.sort(key=lambda x: x[1], reverse=True)
    best_models[ligand] = models[0]  # Top-ranked JSON
    print(f"\n{ligand} ranked models:")
    for file, score in models:
        print(f"  {file}: mean pLDDT = {score:.2f}")

# Copy top .pdb for each ligand to best_ligands/
for ligand, (best_json, best_score) in best_models.items():
    pdb_file = best_json.replace(".json", ".pdb").replace("_scores_", "_unrelaxed_")
    pdb_path = os.path.join(folder, pdb_file)
    if os.path.exists(pdb_path):
        dest = os.path.join(best_folder, pdb_file)
        shutil.copy(pdb_path, dest)
        print(f"✅ Copied {pdb_file} (mean pLDDT = {best_score:.2f}) to best_ligands/")
    else:
        print(f"⚠️ No matching PDB found for {ligand} ({pdb_file})")

print("\nAll top models processed successfully.")



#The script loops through all ligand output files, ranks each model by its mean pLDDT score, selects the highest-confidence model for every ligand, creates a new folder called best_ligands, copies the corresponding .pdb files into that folder, and finally prints a clear summary of the selected top models.
## This is the ouput ligands we're going to use now for autodock
```

```python
# Selecting best ARP ligand models (Arp1–Arp3)
import os, json, shutil
import numpy as np

# Path to your ARP ColabFold output folder
folder = "/Users/derbhladuffy/Desktop/diss_project/annaconda_env_colabfold/arps_output"
best_folder = os.path.join(folder, "best_arps")

# Create the destination folder if it doesn't already exist
os.makedirs(best_folder, exist_ok=True)

results = {}       # Store ranked models for each ARP
best_models = {}   # Store the best model per ARP

# Loop through all JSON files in the ARP output directory
for file in os.listdir(folder):
    if file.endswith(".json") and file.startswith("Arp"):
        ligand = file.split("_")[0]   # "Arp1", "Arp2", "Arp3"
        file_path = os.path.join(folder, file)

        with open(file_path) as f:
            data = json.load(f)

        # Only evaluate models that contain pLDDT scores
        if "plddt" in data:
            mean_plddt = np.mean(data["plddt"])
            results.setdefault(ligand, []).append((file, mean_plddt))

# Rank models for each ARP ligand
for ligand, models in results.items():
    models.sort(key=lambda x: x[1], reverse=True)
    best_models[ligand] = models[0]

    print(f"\n{ligand} ranked models:")
    for file, score in models:
        print(f"  {file}: mean pLDDT = {score:.2f}")

# Copy best models to best_arps/
print("\nCopying highest-confidence ARP models...")
for ligand, (best_json, best_score) in best_models.items():

    # Find corresponding unrelaxed PDB file for docking
    pdb_file = best_json.replace(".json", ".pdb").replace("_scores_", "_unrelaxed_")
    pdb_path = os.path.join(folder, pdb_file)

    if os.path.exists(pdb_path):
        dest = os.path.join(best_folder, pdb_file)
        shutil.copy(pdb_path, dest)
        print(f"✅ Copied {pdb_file} (mean pLDDT = {best_score:.2f})")
    else:
        print(f"⚠️ No matching PDB found for: {ligand}")

print("\nAll ARP ligand models processed.")
```
